Product Details

```python
df2=pd.read_csv(r"C:\Users\r21ec\Downloads\Products.csv")
df2
```

```python
df2.info()
```

```python
df2['ProductKey'].nunique()
```

```python
df2['Unit Cost USD']=df2['Unit Cost USD'].str.replace('$','')

```

```python
df2['Unit Cost USD']=df2['Unit Cost USD'].str.replace(',','')
```

```python
df2['Unit Cost USD']=df2['Unit Cost USD'].astype(float)
```

```python
df2['Unit Cost USD']
```

```python
df2['Unit Price USD']=df2['Unit Price USD'].str.replace('$','')
```

```python
df2['Unit Price USD']=df2['Unit Price USD'].str.replace(',','')
```

```python
df2['Unit Price USD']=df2['Unit Price USD'].astype(float)
```

```python
df2.info()
```

```python
df2
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply MinMax scaling
df2[['Unit Cost USD','Unit Price USD']] = scaler.fit_transform(df2[['Unit Cost USD','Unit Price USD']])

df2

```

```python
df2.describe()
```

```python
df2.to_csv('product.csv',index=False)
```
